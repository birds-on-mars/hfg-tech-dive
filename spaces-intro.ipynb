# Of strange spaces

How are pictures stored in a computer? Pictures are stored in the form of *pixels*. For a gray-scale picture, each pixel is a single number representing the *brightness* of the specific area of the image that is taken by the pixel.

A 3x3 pixel image $x$ might look to the computer something like this:

$ x = \begin{bmatrix} 0.2 & 0.3 & 0.2 \\ 0.5 & 0.4 & 0.1 \\ 0.5 & 0.3 & 0.1 \\ \end{bmatrix}$

... but actually the computer does not care to much for the square layout. The same picture could instead represented in *flattened* form, as a *vector of numbers*:

$ x = \begin{bmatrix} 0.2 \\ 0.3 \\ 0.2 \\ 0.5 \\ 0.4 \\ 0.1 \\ 0.5 \\ 0.3 \\ 0.1 \\ \end{bmatrix}$

As you might vaguely remember from school, a *vector* points somewhere in space.

## but what space?

Let's back up a little and imagine a picture that consists of exactly one pixel. As before, the brightness of the pixel could be represented by a single value. That value can be identified on a scale, ranging from the minimum value (black) to the maximum value (white). Somewhere on this continuum falls our value.




Now image, our picture had two pixels. For each pixel, we put up a continuum, and depending of the brightness of the two pixels, our picture marks a point on a *plane* that is spanned by the two continua.


As we add more dimensions, the space becomes a little overwhelming to imagine. Nevertheless, pictures and other data stored on the hard drive can be regarded as being locations in a high dimensional space.

Vectors can be stretched, rotated or even projected into spaces with higher or lower dimensionality with matrix multiplications. This is the realm of *Linear Algebra*. Machine learning models are such matrix multiplications, with addition of some *non-linearities*. During training, the model is fitted to learn the *desired* transformation of the data vectors.

In the following example, a model $M$ projects some data vector $x$, for example an image of your pet, and projects it into a *one-dimensional space*: a single number that represents the probability, that your pet is a cat. Non-linearities are left out for simplicity.


$ f(x) = x^T \cdot M =  \begin{bmatrix} 0.5 & 0.2 & 0.2 \\ \end{bmatrix} \cdot \begin{bmatrix} 0.1 \\ 0.1 \\ 0.7 \\ \end{bmatrix}  = 0.21 $
